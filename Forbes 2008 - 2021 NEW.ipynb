```python
Financials.name = "Financials"
Tech.name = "Technology"
Petroleum.name = "Petroleum"

industries=[Financials,Tech,Petroleum]
years = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]


def industry_analysis(x_df):
   # get the highest and lowest profits
    x_df.reset_index(inplace=True, drop=True)
    x_df_highest_profits = x_df.loc[x_df.groupby("Year")["Profits"].idxmax()]
    x_df_lowest_profits = x_df.loc[x_df.groupby("Year")["Profits"].idxmin()]

    # Using for loop to plot the profit ranges in pie and bar for all years
    for year in years:
        profit_year = x_df[x_df['Year'] == year]
        ranges = profit_year.groupby('Profit Range')['Profit Range'].count()
        plt.pie(ranges.values,labels=ranges.index.values,autopct=lambda p: '{:.1f}%'.format(round(p)) if p > 0 else '',radius=1,textprops={'fontsize': 7})
        plt.title(f"{x_df.name} Sector Profit Ranges for {year}", pad=5)
        plt.legend(bbox_to_anchor=(-0.15, .5), fontsize=8)
        plt.savefig(f"Output/{x_df.name}_pr{year}_pie.png", bbox_inches='tight')
        plt.show()
        ranges.plot.bar()
        plt.xticks(rotation=45, ha='center')
        plt.ylabel('Number of Companies')
        plt.xlabel('Profit Range')
        plt.savefig(f"Output/{x_df.name}_pr{year}_bar.png", bbox_inches='tight')
        plt.show()    
        info = profit_year.aggregate({'Profits':["mean", "median", "var", "std", "sem", "min", "max"]})    
        print(ranges)
        print(info)
    
def profit_ranges(x_df):
    
    # Creating empty lists for line plot of profit ranges
    loss = []
    less = []
    between = []
    above = []
    
    for year in years:
        profits = x_df[x_df['Year'] == year]
        range = profits.groupby('Profit Range')['Profit Range'].count()
        loss.append(range.iloc[0])
        less.append(range.iloc[1])
        between.append(range.iloc[2])
        above.append(range.iloc[3])

    # Multiline plot for profit ranges 
    plt.plot(years, loss, label = 'Profit Loss/No Profit', linestyle = '-')
    plt.plot(years, less, label = 'Less than $5 billion', linestyle = '--')
    plt.plot(years, between, label = 'Between $5-10 billion', linestyle = 'dotted')
    plt.plot(years, above, label = 'Above $10 billion', linestyle = 'dashdot')
    plt.xticks()
    plt.ylabel('Number of Companies')
    plt.xlabel('Year')
    plt.title(f'{x_df.name} Sector Profit Ranges Over the Years')
    plt.savefig(f"Output/{x_df.name}_Sector_Profit _yoy.png")
    plt.legend(bbox_to_anchor=(-0.15, .5), fontsize=8)
    plt.show()
    
def profit_min_max(x_df):
    
    # Finding the profit min from all years
    highest_profits = x_df.loc[x_df.groupby("Year")["Profits"].idxmax()]
    highest_profits
    lowest_profits = x_df.loc[x_df.groupby("Year")["Profits"].idxmin()]
    lowest_profits
    
    # Assigning variables for x and y values
    year1 = highest_profits['Year']
    highs = highest_profits['Profits']
    year2 = lowest_profits['Year']
    lows = lowest_profits['Profits']
    
    # Plotting multiline for the min and max of profits over the years
    plt.plot(year1, highs, label = 'Highest Profit', linestyle = '-')
    plt.plot(year2, lows, label = 'Lowest Profit', linestyle = '--')
    plt.ylabel('Profits (billions)')
    plt.xlabel('Year')
    plt.xticks([x for x in year1], rotation=45)
    plt.title(f'{x_df.name} Highest and Lowest Profits over the Years')
    plt.legend(bbox_to_anchor=(-0.15, .5), fontsize=8)
    plt.savefig(f"Output/{x_df.name}_pr_high_low.png",  bbox_inches='tight')
    plt.show()
    
```

```python
for x in industries:
    industry_analysis(x)
    profit_ranges(x)
    profit_min_max(x)
```

The above plots show that during the 2008 Housing Crisis, there weren't any companies in the financial sector that made over 5 billion and the percentage of companies that experienced no profit or a profit loss doubled in 2009.  While in 2020, during the Covid19 Pandemic, it shows a decrease in companies with less than $5 billion in profit and an increase for companies that made more than 5 billion.  It appears that the 2008 Housing Crisis had a negative impact on the financial sector, while the Covid19 Pandemic had a positive impact.

The above plots show that during the 2008 Housing Crisis, the technology sector did not have a significant impact.  Although more companies showed a profit loss, there was also an increase in companies that made over 5 billion.  While in 2020, during the Covid19 Pandemic, it shows only a few companies with profit loss and more companies with over 5 billion in profit.  The maximum profit for technology also mostly increased over the years and the minimum profit mostly decreased over the years.  It appears that neither the 2008 Housing Crisis or Covid19 Pandemic, impacted the technology sector much.

In the above plots for the petroleum sector, it shows that companies experienced a decrease in profit during the 2008 Housing Crisis, particularly in years 2009 and 2010. During the Covid19 Pandemic, it also shows that companies experienced a decrease in profit.

The profit min and max multiline plot shows a steep decline in a comparison of the financial sector's maximum and minimum profits for 2009, while there was a jump in maximum and minimum profits in 2020.
